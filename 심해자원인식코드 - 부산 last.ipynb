```python
import time as ti
import winsound as win
import urllib.request
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from tqdm.notebook import tqdm
import csv
from math import pi, cos, sin
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from random import randint
import serial
import easygui
import tkinter as tk
import cv2
import datetime
import os

dic2={}
list1=[]
count=0
r=0
num=[]
w=0
list2=[]
rock=0
s=0
    
# 실행하기 전 port가 맞는지 확인!
PORT=easygui.enterbox('포트가 COM 몇 번에 연결되어있나요? ex) COM4')
BaudRate = 9600
ARD=serial.Serial(PORT,BaudRate)

easygui.msgbox('수중 로봇의 장애물 회피 및 최적 경로 탐구')


ask=easygui.buttonbox('''eastsea_depth 파일이 코드와 동일 폴더에 있습니까?
파일이 있으면 y를, 없으면 n을 입력해주세요.''',choices =['y','n'])


if ask=='n':
    path="C:\chromedriver.exe"
    link='https://topex.ucsd.edu/cgi-bin/get_srtm30.cgi'
    height=[]

    driver1=webdriver.Chrome(path)
    driver1.get(link)

    elem1 = driver1.find_element_by_name("east")
    elem2 = driver1.find_element_by_name("west")
    elem3 = driver1.find_element_by_name("south")
    elem4 = driver1.find_element_by_name("north")
    elem1.clear()
    elem2.clear()
    elem3.clear()
    elem4.clear()
    elem1.send_keys("136.01")
    elem2.send_keys("129.71")
    elem3.send_keys("37.495")
    elem4.send_keys("39.09")
    elem4.submit()

    html=driver1.page_source
    soup=BeautifulSoup(html,'html.parser')
    sent1=str(soup)[85:-20].split('\n')
    for i in range(len(sent1)-1):
        height.append(sent1[i].split('\t'))

    driver1.close()

    driver2=webdriver.Chrome(path)
    driver2.get(link)

    elem1 = driver2.find_element_by_name("east")
    elem2 = driver2.find_element_by_name("west")
    elem3 = driver2.find_element_by_name("south")
    elem4 = driver2.find_element_by_name("north")
    elem1.clear()
    elem2.clear()
    elem3.clear()
    elem4.clear()
    elem1.send_keys("136.01")
    elem2.send_keys("129.71")
    elem3.send_keys("35.73")
    elem4.send_keys("37.505")
    elem4.submit()

    html=driver2.page_source
    soup=BeautifulSoup(html,'html.parser')
    sent2=str(soup)[85:-20].split('\n')
    for i in range(len(sent2)-1):
        height.append(sent2[i].split('\t'))

    driver2.close()

    fout=open('eastsea_depth.csv','w',encoding='utf-8',newline='')
    csvW=csv.writer(fout)
    csvW.writerow(['longitude','latitude','depth'])
    csvW.writerows(height)

    fout.close()


def Ardread():
    try:
        LINE=ARD.readline()
        code0=int(LINE.decode('utf-8')[:-2].split(',')[0])
        code1=int(LINE.decode('utf-8')[:-2].split(',')[1])
        code2=float(LINE.decode('utf-8')[:-2].split(',')[2])
        return [code0,code1,code2]
    except:
        pass

def Ardcount():
    while True:
        s_time=ti.time()
        ans=Ardread()
        f_time=ti.time()
        timer=f_time-s_time
        if timer>0.1:
            return ans
            break
        else:
            continue

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]        
    
while True:
    boat=float(easygui.enterbox('잠수함의 속력(m/s)은? (최소 : 3) :'))
    if boat<3:
        easygui.msgbox('3보다 작습니다. 다시 입력하세요.')
        continue
    else:
        break
r=float(easygui.enterbox('모형에서의 지점 간의 거리(cm)를 입력하세요. :'))

class Graph:
    def __init__(self):
        self.node = set()   # node에 있는 값과 거리들을 리셋
        self.edges = {}
        self.distances = {}
 
    def add_node(self, value):
        self.nodes.add(value)
 
    def add(self,from_node,to_node,velocity):
        global boat
        self._add_edge(from_node, to_node, 1/(boat+velocity))   # 한 node에서 이웃한 node로 갈 때의 해류의 속도를 받아 속도의 역수인 거리 꼴로 저장
        self._add_edge(to_node, from_node, 1/(boat-velocity))   # 반대로 이동하는 경우
        dic2[from_node,to_node]=velocity     # 두 node 사이의 해류 속도를 dic2에 저장
        dic2[to_node,from_node]=-velocity    # 반대의 경우도 저장

    def add2(self,from_node,to_node,velocity):
        global boat
        self._add_edge(from_node, to_node, 2/(boat-velocity))   # 장애물의 경우에 사용
        self._add_edge(to_node, from_node, 2/(boat-velocity))
        dic2[from_node,to_node]=-velocity
        dic2[to_node,from_node]=-velocity
        list1.append([from_node,to_node,velocity])

    def add3(self,from_node,to_node,velocity):
        global boat
        self._add_edge(from_node, to_node, 2/(boat-velocity))   # 암초의 경우에 사용
        self._add_edge(to_node, from_node, 2/(boat-velocity))
        dic2[from_node,to_node]=-velocity
        dic2[to_node,from_node]=-velocity
        list2.append([from_node,to_node,velocity])
 
    def _add_edge(self, from_node, to_node, distance):
        self.edges.setdefault(from_node, [])
        self.edges[from_node].append(to_node)
        self.distances[(from_node, to_node)] = distance


def dijkstra(graph, initial_node):
    visited = {initial_node: 0}
    path = {}
 
    nodes = set(graph.nodes)
 
    while nodes:
        min_node = None
        for node in nodes:
            if node in visited:
                if min_node is None:                                   # 제일 처음 찾은 경로를 최소 경로로 일단 세팅
                    min_node = node
                elif visited[node] < visited[min_node]:                # 원래 찾았던 경로보다 이 경로가 더 짧게 걸릴 경우 최소 경로를 그 경로로 교체
                    min_node = node
        if min_node is None:
            break
 
        nodes.remove(min_node)
        cur_wt = visited[min_node]
 
        for edge in graph.edges[min_node]:
            wt = cur_wt + graph.distances[(min_node, edge)]
            if edge not in visited or wt < visited[edge]:
                visited[edge] = wt
                path[edge] = min_node
    return visited, path
 
def shortest_path(graph, initial_node, goal_node):                  # 처음과 끝 지점을 받고, dijkstra 함수에서 받은 경로들을 비교하여 최단 경로 찾기
    distance, path = dijkstra(graph, initial_node)
    route = [goal_node]
    while goal_node != initial_node:
        route.append(path[goal_node])
        goal_node = path[goal_node]
 
    route.reverse()
    return route


dictot={}
temp1=[i for i in range(467004,489573,806)]
temp2=[i for i in range(128703,140689,705)]
t_day=int(ti.strftime('%Y%m%d'))
t_hour=int(ti.strftime('%H'))
checkw=0

while True:
    easygui.msgbox('기존의 해류 파일 eastsea_current 가 코드와 동일 경로에 있는지 확인한 뒤 실행해주세요!')
    question=easygui.buttonbox('기존의 파일을 사용하시려면 s 를, 새로운 현재 시각의 해류 데이터를 사용하시려면 t 를, 새로운 미래의 예측 해류 데이터를 사용하시려면 f 를 눌러주세요. : ',choices=['s','t','f'])
    if question=='t':
        day=t_day
        hour=t_hour
        checkw=1
        break
    elif question=='f':
        questday=int(esaygui.buttonbox('오늘이면 0, 내일이면 1, 모레이면 2 를 입력해주세요. : ',choices=['0','1','2']))
        questhour=int(easygui.enterbox('몇시인지 입력해주세요. (모레 22시까지 가능) : '))
        day=t_day+questday
        hour=questhour
        checkw=1
        break
    elif question=='s':
        break
    else:
        easygui.msgbox('제대로 입력하세요.')
        continue
    
    
if checkw==1:
    fout=open('eastsea_current.csv','w',encoding='utf-8',newline='')
    csvW=csv.writer(fout)
    csvW.writerow(['start','end','current'])
    mm=[]

    for in1 in tqdm(range(18),desc='전체 진행률'):
        for in2 in tqdm(range(29),desc='부분 진행률'):
            num1=str(temp1[in2]/3600)
            num2=str(temp2[in1]/3600)
            url='http://www.khoa.go.kr/oceangrid/grid/api/romsCurrent/search.do?ServiceKey=unbe8DC11rWidPjbdfDIQ==&ObsLon='+num1+'&ObsLat='+num2+'&ResultType=json'
            byte_data = urllib.request.urlopen(url).read()

            soup=BeautifulSoup(byte_data,'html.parser')

            diccur={}
            a = str(soup)[20:].split('},{')[:-1]
            dilist=[]
            holist=[]
            dalist=[]
            splist=[]
            for i in range(len(a)):
                dilist.append(float(a[i].split(',')[0][18:-1]))
                holist.append(int(a[i].split(',')[1][8:-1]))
                dalist.append(int(a[i].split(',')[2][8:-1]))
                splist.append(float(a[i].split(',')[3][17:-1]))

            for k in range(len(a)):
                diccur[dalist[k],holist[k]]=[splist[k],dilist[k]]
            dictot[29*in1+in2+1]=diccur[day,hour]


    for ay in range(18):
        for ax in range(1,29):
            mm.append([ax+ay*29,ax+ay*29+1,dictot[ax+ay*29][0]*cos(dictot[ax+ay*29][1]*pi/180)])
            mm.append([ax+ay*29+1,ax+ay*29,dictot[ax+ay*29+1][0]*(-1)*cos(dictot[ax+ay*29+1][1]*pi/180)])


    for ay in range(17):
        for ax in range(29):
            mm.append([ax+ay*29+1,ax+ay*29+30,dictot[ax+ay*29+1][0]*sin(dictot[ax+ay*29+1][1]*pi/180)])
            mm.append([ax+ay*29+30,ax+ay*29+1,dictot[ax+ay*29+30][0]*(-1)*sin(dictot[ax+ay*29+30][1]*pi/180)])

    csvW.writerows(mm)
    fout.close()

file=np.genfromtxt("eastsea_current.csv",dtype=float,delimiter=',',skip_header=1)
file2=np.genfromtxt("eastsea_depth.csv",dtype=float,delimiter=',',skip_header=1)

g = Graph()
dic_depth={}

# 경도 756개 -> 29개, 위도 405개 -> 18개
longitude=np.array([file2[k][0] for k in range(756)])
latitude=np.array([file2[k][1] for k in range(0,len(file2),756)])
dic_longi={}
dic_lati={}
dic_height={}
dicall={}
for r1 in range(len(temp1)):
    dic_longi[r1]=find_nearest(longitude,temp1[r1]/3600)
for r2 in range(len(temp2)):
    dic_lati[r2]=find_nearest(latitude,temp2[r2]/3600)
for r3 in range(len(file2)):
    dic_height[file2[r3][0],file2[r3][1]]=file2[r3][2]
for p in range(18):
    for q in range(29):
        dicall[29*p+q+1]=(int(dic_height[dic_longi[q],dic_lati[p]])+49)//100
zlist=[]
        
for i in range(len(file)):
    for z in range(0,1-dicall[int(file[i][0])]):
        if -dicall[int(file[i][1])]>=z:
            g.add(int(file[i][0])+1000*z,int(file[i][1])+1000*z,file[i][2])         #g.add는 전부 가중치 관련 부분이므로 협의 필요
            zlist.append(int(file[i][0])+1000*z)

for j in range(1,523):
    for k in range(0,-dicall[j]):
        g.add(j+k*1000,j+(k+1)*1000,-k/10)
        g.add(j+(k+1)*1000,j+k*1000,k/10)
        
g.nodes = set(zlist)      # 지점 설정

dic1={}
for poy in range(18):
    for pox in range(29):
        for poz in range(0,1-dicall[29*poy+pox+1]):
            dic1[1000*poz+29*poy+pox+1]=[pox,poy,-poz]

while True:
    while True:
        a=int(easygui.enterbox('출발점을 입력하세요. :'))
        b=int(easygui.enterbox('도착점을 입력하세요. :'))
        if a and b in zlist:
            break
        else:
            easygui.msgbox('존재하지 않는 지점입니다. 제대로 입력하세요.')
            continue
    
    while True:
        k=easygui.buttonbox('현재 '+str(count)+'개의 장애물과 '+str(rock)+'개의 암초가 입력되어 있습니다. 장애물을 추가하려면 + 를, 더 추가할 장애물이 없다면 - 를 입력하세요. :',choices=['+','-'])
        if k=='+':
            while True:
                easygui.msgbox('어느 두 인접한 점 사이인지 입력하세요.')
                x=int(easygui.enterbox('첫번째 점:'))
                y=int(easygui.enterbox('두번째 점:'))
                if (x,y) in dic2.keys():
                    pass
                else:
                    easygui.msgbox('인접한 두 점이 아닙니다. 다시 입력하세요.')
                    continue
                while True:
                    z=float(easygui.enterbox('피해 정도를 입력하세요. (최소 : 1, 최대 : 배의 속력 미만) :'))    # 해류 속도의 3분의 1 정도에서 피해 정도를 감수하고 갈지 말지가 결정되는 듯
                    if z<1:
                        easygui.msgbox('너무 작습니다. 범위에 맞게 입력하세요.')                              # 해류 속도에 비례하게 비율로 생각하여 피해 정도를 세팅
                        continue
                    elif z>=boat:
                        easygui.msgbox('너무 큽니다. 범위에 맞게 입력하세요.')
                        continue
                    else:
                        break
                g.add2(x,y,z)
                count+=1
                break
        elif k=='-':
            break
        else:
            easygui.msgbox('제대로 입력하세요.')
            continue
            
    s_path=shortest_path(g,a,b)
    
    time=0

    for k in range(len(s_path)-1): 
        time+=(200000/3)/(boat+dic2[s_path[k],s_path[k+1]])

    easygui.msgbox('이동 거리 :'+str(int((200/3)*(len(s_path)-1)))+'km\n'+'\n'+'걸리는 시간 :'+str(int(time//86400))+'일'+str(int((time-(time//86400)*86400)//3600))+'시간'+str(int((time-(time//3600)*3600)//60))+'분'+str(int(time-(time//3600)*3600-((time-(time//3600)*3600)//60)*60))+'초')
    
    fig=plt.figure(figsize=(7,7))
    ax=fig.gca(projection='3d')
    #ax2=fig.gca(projection='3d')
    
    x=[]
    y=[]
    z=[]
    listk=[]
    listp=[]
    x2=[]
    y2=[]
    z2=[]
    
    for k in range(len(s_path)):
        x.append(dic1[s_path[k]][0])
        y.append(dic1[s_path[k]][1])
        z.append(dic1[s_path[k]][2])

    for i in range(len(x)):
        listk.append(np.array([x[i],y[i],z[i]]))

    for j in range(1,len(x)):
        check=listk[j]-listk[j-1]
        listp.append('x' if check[0]!=0 else 'y' if check[1]!=0 else 'z')
    
    for n in range(1,523):
        if n in dic1.keys():
            x2.append(dic1[n][0])
            y2.append(dic1[n][1])
            z2.append(dicall[n])
        else:
            x2.append((n-1)%29)
            y2.append((n-1)//29)
            z2.append(dicall[n])

    listq=[i for i in enumerate(listp)]

    ax.plot(x,y,z,'r-')
    #ax2.plot(x2,y2,z2,'ok')
    
    #plt.show()
    plt.savefig('quickpath.png')
    
    reply=easygui.msgbox(msg=(str(s_path)+'\n'+str(listq)+'\n'),image='quickpath.png')

    while True:
        s=Ardcount()
        print(s)
        if s[2]<=r:   
            #for u in range(2):
                #win.Beep(5000,750)
                #win.Beep(3000,750)
            win.Beep(5000,1000)
            easygui.msgbox('암초가 발견되었습니다. 경로 재탐색 및 암석 촬영을 시작합니다.')
            
            video_capture = cv2.VideoCapture(0)
            
            grabbed, frame = video_capture.read()
            cv2.imshow('Original Video', frame)

            file = datetime.datetime.now().strftime("%Y%m%d_%H%M%S%f") + '.png'
            #cv2.imwrite(os.path.join("c:/testimage/subimage/", str(file)), frame)
            cv2.imwrite(os.path.join("c:/testimage/subimage2/", str(file)), frame)
            #easygui.msgbox(file, ' saved',image='c:/testimage/subimage/'+str(file))
            easygui.msgbox('image saved')

            video_capture.release()
            cv2.destroyAllWindows()
            
            while True:
                while True:
                    p1=int(easygui.enterbox('현재 위치가 어딥니까? :'))
                    if p1 in s_path: break
                    else:
                        easygui.msgbox('경로에 없는 위치입니다. 제대로 입력하세요.')
                        continue
                if p1==b:
                    w=1
                    break
                p2=s_path[s_path.index(p1)+1]
                p_list=[p1,p2+1,p2+29,p2-29,p2+1000,p2-1000]
                
                for p in range(6):
                    if (p_list[p],p2) in dic2.keys():
                        g.add3(p_list[p],p2,boat/2)
                        g.add3(p2,p_list[p],boat/2)

                rock+=1
                easygui.msgbox('암초의 정보가 입력되었습니다. 다시 현재 위치와 도착지를 입력해주세요.')
                break
            break
        else:
            continue
    if w==1:
        easygui.msgbox('무사히 도착하신 것을 축하드립니다. 지금부터 발견한 암석에 대한 확인을 시작하겠습니다.')
        break
    else:
        continue

plt.cla()
plt.clf()
        
from PIL import Image
import os, glob, numpy as np
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 

caltech_dir = "C:/testimage"
categories = ["man","gas","amcho"]
nb_classes = len(categories)

image_w = 64
image_h = 64

pixels = image_h * image_w * 3

X = []
y = []

for idx, cat in enumerate(categories):
    
    #one-hot 돌리기.
    label = [0 for i in range(nb_classes)]
    label[idx] = 1

    image_dir = caltech_dir + "/" + cat
    print(image_dir)
    files = glob.glob(image_dir+"/*.png")
    print(cat, " 파일 길이 : ", len(files))
    for i, f in enumerate(files):
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((image_w, image_h))
        data = np.asarray(img)

        X.append(data)
        y.append(label)

        if i % 700 == 0:
            print(cat, " : ", f)

X = np.array(X)
y = np.array(y)
#1 0 0 0 이면 airplanes
#0 1 0 0 이면 buddha 이런식


X_train, X_test, y_train, y_test = train_test_split(X, y)
xy = (X_train, X_test, y_train, y_test)
np.save("./multi_image_data.npy", xy)

print("ok", len(y))

import os, glob, numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import keras.backend.tensorflow_backend as K

import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
session = tf.Session(config=config)

X_train, X_test, y_train, y_test = np.load('./multi_image_data.npy', allow_pickle=True)
#print(X_train.shape)
#print(X_train.shape[0])

categories = ["man","gas","amcho"]
nb_classes = len(categories)

#일반화
X_train = X_train.astype(float) / 255
X_test = X_test.astype(float) / 255

with K.tf_ops.device('/device:GPU:0'):
    model = Sequential()
    model.add(Conv2D(32, (3,3), padding="same", input_shape=X_train.shape[1:], activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))
    
    model.add(Conv2D(64, (3,3), padding="same", activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))
    
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(nb_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model_dir = './model'
    
    if not os.path.exists(model_dir):
        os.mkdir(model_dir)
    
    model_path = model_dir + '/multi_img_classification.model'
    checkpoint = ModelCheckpoint(filepath=model_path , monitor='val_loss', verbose=1, save_best_only=True)
    early_stopping = EarlyStopping(monitor='val_loss', patience=6)
    
model.summary()

#데이터셋이 적어서 validation을 그냥 test 데이터로 했습니다. 
#데이터셋이 충분하시면 이렇게 하시지 마시고 validation_split=0.2 이렇게 하셔서 테스트 셋으로 나누시길 권장합니다.
history = model.fit(X_train, y_train, batch_size=32, epochs=50, validation_data=(X_test, y_test), callbacks=[checkpoint, early_stopping])

print("정확도 : %.4f" % (model.evaluate(X_test, y_test)[1]))

y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len, y_vloss, marker='.', c='red', label='val_set_loss')
plt.plot(x_len, y_loss, marker='.', c='blue', label='train_set_oss')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()
plt.show()

from PIL import Image
import os, glob, numpy as np
from keras.models import load_model

caltech_dir = "c:/testimage/subimage"
image_w = 64
image_h = 64

pixels = image_h * image_w * 3

X = []
filenames = []
files = glob.glob(caltech_dir+"/*.png")
for i, f in enumerate(files):
    img = Image.open(f)
    img = img.convert("RGB")
    img = img.resize((image_w, image_h))
    data = np.asarray(img)
    filenames.append(f)
    X.append(data)

X = np.array(X)
model = load_model('./model/multi_img_classification.model')

prediction = model.predict(X)
np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})
cnt = 0

#이 비교는 그냥 파일들이 있으면 해당 파일과 비교. 카테고리와 함께 비교해서 진행하는 것은 _4 파일.
for i in prediction:
    pre_ans = i.argmax()  # 예측 레이블
    #print(i)
    #print(pre_ans)
    pre_ans_str = ''
    if pre_ans == 0: pre_ans_str = "망간단괴"
    elif pre_ans == 1: pre_ans_str = "가스하이드레이트"
    elif pre_ans == 2: pre_ans_str = "암초"
    #else: pre_ans_str = "게"
    if i[0] >= 0.8 : print("해당 "+filenames[cnt].split("\\")[1]+"이미지는 "+pre_ans_str+"로 추정됩니다.")
    print('')
    if i[1] >= 0.8: print("해당 "+filenames[cnt].split("\\")[1]+"이미지는 "+pre_ans_str+"으로 추정됩니다.")
    if i[2] >= 0.8: print("해당 "+filenames[cnt].split("\\")[1]+"이미지는 "+pre_ans_str+"로 추정됩니다.")
    #if i[3] >= 0.8: print("해당 "+filenames[cnt].split("\\")[1]+"이미지는 "+pre_ans_str+"로 추정됩니다.")
    cnt += 1
    # print(i.argmax()) #얘가 레이블 [1. 0. 0.] 이런식으로 되어 있는 것을 숫자로 바꿔주는 것.
    # 즉 얘랑, 나중에 카테고리 데이터 불러와서 카테고리랑 비교를 해서 같으면 맞는거고, 아니면 틀린거로 취급하면 된다.
    # 이걸 한 것은 _4.py에.
    
```
